```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = 11,7
```

```python
df_2015 = pd.read_csv('2015.csv')
df_2015['Year'] = 2015

df_2016 = pd.read_csv('2016.csv')
df_2016['Year'] = 2016

df_2017 = pd.read_csv('2017.csv')
df_2017['Year'] = 2017

df_2018 = pd.read_csv('2018.csv')
df_2018['Year'] = 2018

df_2019 = pd.read_csv('2019.csv')
df_2019['Year'] = 2019
```

```python
df_2015_filtered = df_2015.loc[:, ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Year']]
df_2015_filtered.head()
```

```python
df_2016_filtered = df_2016.loc[:, ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Year']]
df_2016_filtered.columns = ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Year']
df_2016_filtered.head()
```

```python
df_2017_filtered = df_2017.loc[:, ['Country', 'Happiness.Rank', 'Happiness.Score', 'Economy..GDP.per.Capita.', 'Family', 'Health..Life.Expectancy.', 'Freedom', 'Trust..Government.Corruption.', 'Year']]
df_2017_filtered.columns = ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Year']
df_2017_filtered.head()
```

```python
df_2018_filtered = df_2018.loc[:, ['Country or region', 'Overall rank', 'Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Perceptions of corruption', 'Year']]
df_2018_filtered.columns = ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Year']
df_2018_filtered.head()
```

```python
df_2019_filtered = df_2019.loc[:, ['Country or region', 'Overall rank', 'Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Perceptions of corruption', 'Year']]
df_2019_filtered.columns = ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Year']
df_2019_filtered.head()
```

```python
df = pd.concat([df_2015_filtered, df_2016_filtered, df_2017_filtered, df_2018_filtered, df_2019_filtered])
df = df.merge(df_2015[['Country','Region']], on='Country')
df.head()
```

```python
df.info()
```

```python
dynamics_by_year = df.groupby('Year').agg(['mean', 'median'])['Happiness Score']
dynamics_by_year
```

```python
#построим общую аналитику
#показатели ВВП и свободы жизненного выбора вероятно обладают наимененее выраженной зависимостью
sns.pairplot(df)
```

```python
# в целом ожидаем, что люди более счастливы, если ожидают большую продолжительность жизни
sns.regplot(data=df, x='Health (Life Expectancy)', y='Happiness Score')
plt.title('Корреляция между счастьем и продолжительностью жизни')
```

```python
# при возможном диапазоне оценок от 1 до 10, что динамика среднего значения счастья год к году меняется не значительно
dynamics_by_year.plot(title='Динамика счастья',
                             xlabel='Year',
                             ylabel='mean, median')
plt.xticks(dynamics_by_year.index)
plt.ylim([5, 6])
```

```python
#при анализе экономики видим выброс в 2018 году
#при этом с 2018 года медиана смещается ближе к середине
#интересное наблюдение, что при росте показателя в 2017 году наблюдается снижение общего показателя счастья, пусть и не значительно
sns.boxplot(data=df, x="Year", y="Economy (GDP per Capita)")
```

```python
# при 10-бальной шкале большое часть попадает в середину, что говорит о том, что довольно мало ярко выраженных счастливых стран
sns.histplot(df['Happiness Score'])
plt.title('Распределение оценок счастья')
```

```python
# присвоим условные категории счастья странам
happy = 6.67
not_happy = 3.33
df.loc[df['Happiness Score'] >= happy, 'Happiness_category'] = 'Happy'
df.loc[df['Happiness Score'] < not_happy, 'Happiness_category'] = 'Unhappy'
df['Happiness_category'] = df['Happiness_category'].fillna('Middle_happy')
grouped_by_happy = df.pivot_table(
                                  index='Region',
                                  columns='Happiness_category',
                                  values='Year',
                                  aggfunc='count')
grouped_by_happy
```

```python
#видим, что в Северной Америке и Австралии только условно счастливые страны
#несчастные страны в большой степени имеются на Ближнем Востоке и Северной Африке, в Южной Азии и Чёрной Африке
sns.heatmap(grouped_by_happy)
```

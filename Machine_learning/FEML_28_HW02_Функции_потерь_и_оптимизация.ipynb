# **2. Самостоятельно реализуйте логистическую регрессию, без использования метода LogisticRegression из библиотеки. Можете использовать библиотеки pandas, numpy, math для реализации. Оформите в виде функции. *Оформите в виде класса с методами.**

```python
#Добавляем константный признак для ускорение реализации
X = np.c_[np.ones(len(X)), X]
X
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def logloss(y, y_proba):
    logloss_1 = np.sum(np.log(y_proba[y == 1] + 1e-30))
    logloss_0 = np.sum(np.log(1 - y_proba[y == 0] + 1e-30))
    logloss_total = -(logloss_0 + logloss_1) / len(y)
    return logloss_total


def gr_logloss(X, W, y):
    y_proba = sigmoid(X @ W)
    grad = X.T @ (y_proba - y)
    return grad
```

```python
import pandas as pd
results = pd.DataFrame(columns=['Метод', 'Метрика', 'Итерация'])
```

# **3. Реализуйте метод градиентного спуска. Обучите логистическую регрессию этим методом. Выберете и посчитайте метрику качества. Метрика должна быть одинакова для всех пунктов домашнего задания. Для упрощения сравнения выберете только одну метрику.**

```python
%%time
# установка минимального значения, на которое должны изменяться веса
eps = 0.00001

# первоначальная точка
np.random.seed(43)
W = np.random.randn(X.shape[1])

# размер шага (learning rate)
learning_rate = 0.01

next_W = W

# количество итераций
n = 200
for i in range(n):
    cur_W = next_W

    # движение в негативную сторону вычисляемого градиента
    next_W = cur_W - learning_rate * gr_logloss(X, W, y)

    # остановка когда достигнута необходимая степень точности
    if np.linalg.norm(cur_W - next_W) <= eps:
        break

    #print(f"Итерация: {i+1}")
    # print(f"Текущая точка {cur_W}| Следующая точка {next_W}")
    y_proba = sigmoid(X @ next_W)
    y_class = np.where(y_proba >= 0.5, 1, 0)
    #accuracy = (y_class == y).sum() / len(y)
    accuracy = accuracy_score(y, y_class)
    #print(f"Logloss {logloss(y, y_proba)}")
    #print(f"Accuracy {accuracy}")
    #print("--------------------------------------------------------")
    results.loc[len(results.index)] = ['Gradient descent', accuracy, i+1]
```

```python
%%time
# установка минимального значения, на которое должны изменяться веса
eps = 0.00001

# первоначальная точка
np.random.seed(43)
W = np.random.randn(X.shape[1])

# размер шага (learning rate)
learning_rate = 0.01

next_W = W

#кол-во точек для Mini-Batch
amount = 10

# количество итераций
n = 200
for i in range(n):
    cur_W = next_W
    idx = np.random.choice(len(X), size=amount, replace=False)

    # движение в негативную сторону вычисляемого градиента
    next_W = cur_W - learning_rate * gr_logloss(X[idx], W, y[idx])

    # остановка когда достигнута необходимая степень точности
    if np.linalg.norm(cur_W - next_W) <= eps:
        break

    #print(f"Итерация: {i}")
    # print(f"Текущая точка {cur_W}| Следующая точка {next_W}")
    y_proba = sigmoid(X @ next_W)
    y_class = np.where(y_proba >= 0.5, 1, 0)
    accuracy = (y_class == y).sum() / len(y)
    #accuracy = accuracy_score(y, y_class)
    #print(f"Logloss {logloss(y, y_proba)}")
    #print(f"Accuracy {accuracy}")
    #print("--------------------------------------------------------")
    results.loc[len(results.index)] = ['Mini-Batch', accuracy, i+1]
```

# **4. Повторите п. 3 для метода скользящего среднего (Root Mean Square Propagation, RMSProp).**

```python
%%time
k = 200
lr = 0.01
eps_1 = 0.00001

# первоначальная точка
np.random.seed(43)
W = np.random.randn(X.shape[1])
weights_0, weights_1, weights_2, weights_3, weights_4, = [W[0]], [W[1]], [W[2]], [W[3]], [W[4]]

alphas = [0, 0, 0, 0, 0]
eps = 10e-7
rho = 0.9
amount = 10

for i in range(k):
    idx = np.random.choice(len(X), size=amount, replace=False)
    grad = gr_logloss(X[idx], W, y[idx])

    alphas[0] = rho * alphas[0] + (1 - rho) * grad[0] ** 2
    alphas[1] = rho * alphas[1] + (1 - rho) * grad[1] ** 2
    alphas[2] = rho * alphas[2] + (1 - rho) * grad[2] ** 2
    alphas[3] = rho * alphas[3] + (1 - rho) * grad[3] ** 2
    alphas[4] = rho * alphas[4] + (1 - rho) * grad[4] ** 2

    lr_k_0 = lr / np.sqrt(alphas[0] + eps)
    lr_k_1 = lr / np.sqrt(alphas[1] + eps)
    lr_k_2 = lr / np.sqrt(alphas[2] + eps)
    lr_k_3 = lr / np.sqrt(alphas[3] + eps)
    lr_k_4 = lr / np.sqrt(alphas[4] + eps)

    new_w_0 = weights_0[-1] - lr_k_0 * grad[0]
    new_w_1 = weights_1[-1] - lr_k_1 * grad[1]
    new_w_2 = weights_2[-1] - lr_k_2 * grad[1]
    new_w_3 = weights_3[-1] - lr_k_3 * grad[1]
    new_w_4 = weights_4[-1] - lr_k_4 * grad[1]

    weights_0.append(new_w_0)
    weights_1.append(new_w_1)
    weights_2.append(new_w_2)
    weights_3.append(new_w_3)
    weights_4.append(new_w_4)
    prev_W = np.array([weights_0[-2], weights_1[-2], weights_2[-2], weights_3[-2], weights_4[-2]])
    curent_W = np.array([weights_0[-1], weights_1[-1], weights_2[-1], weights_3[-1], weights_4[-1]])

    # остановка когда достигнута необходимая степень точности
    if np.linalg.norm(prev_W - curent_W) <= eps_1:
        break

    #print(f"Итерация: {i+1}")
    # print(f"Текущая точка {cur_W}| Следующая точка {next_W}")
    y_proba = sigmoid(X @ curent_W)
    y_class = np.where(y_proba >= 0.5, 1, 0)
    accuracy = (y_class == y).sum() / len(y)
    #accuracy = accuracy_score(y, y_class)
    #print(f"Logloss {logloss(y, y_proba)}")
    #print(f"Accuracy {accuracy}")
    #print("--------------------------------------------------------")
    results.loc[len(results.index)] = ['RMS-Prop', accuracy, i+1]
```

# **5. Повторите п. 3 для ускоренного по Нестерову метода адаптивной оценки моментов (Nesterov–accelerated Adaptive Moment Estimation, Nadam).**

```python
%%time
k = 200
lr = 0.01
eps_1 = 0.00001

# первоначальная точка
np.random.seed(43)
W = np.random.randn(X.shape[1])
weights_0, weights_1, weights_2, weights_3, weights_4, = [W[0]], [W[1]], [W[2]], [W[3]], [W[4]]

alphas = [0, 0, 0, 0, 0]
velocity = [0, 0, 0, 0, 0]
eps = 10e-7

b1 = 0.5
b2 = 0.9

for i in range(k):
    idx = np.random.choice(len(X), size=amount, replace=False)
    grad = gr_logloss(X[idx], W, y[idx])

    velocity[0] = b1 * velocity[0] + grad[0]
    velocity[1] = b1 * velocity[1] + grad[1]
    velocity[2] = b1 * velocity[2] + grad[2]
    velocity[3] = b1 * velocity[3] + grad[3]
    velocity[4] = b1 * velocity[4] + grad[4]

    alphas[0] = b2 * alphas[0] + (1 - b2) * grad[0] ** 2
    alphas[1] = b2 * alphas[1] + (1 - b2) * grad[1] ** 2
    alphas[2] = b2 * alphas[2] + (1 - b2) * grad[2] ** 2
    alphas[3] = b2 * alphas[3] + (1 - b2) * grad[3] ** 2
    alphas[4] = b2 * alphas[4] + (1 - b2) * grad[4] ** 2

    new_w_0 = weights_0[-1] - (lr * velocity[0]) / np.sqrt(alphas[0] + eps)
    new_w_1 = weights_1[-1] - (lr * velocity[1]) / np.sqrt(alphas[1] + eps)
    new_w_2 = weights_1[-1] - (lr * velocity[2]) / np.sqrt(alphas[2] + eps)
    new_w_3 = weights_1[-1] - (lr * velocity[3]) / np.sqrt(alphas[3] + eps)
    new_w_4 = weights_1[-1] - (lr * velocity[4]) / np.sqrt(alphas[4] + eps)

    weights_0.append(new_w_0)
    weights_1.append(new_w_1)
    weights_2.append(new_w_2)
    weights_3.append(new_w_3)
    weights_4.append(new_w_4)
    prev_W_1 = np.array([weights_0[-2], weights_1[-2], weights_2[-2], weights_3[-2], weights_4[-2]])
    curent_W_1 = np.array([weights_0[-1], weights_1[-1], weights_2[-1], weights_3[-1], weights_4[-1]])

    # остановка когда достигнута необходимая степень точности
    if np.linalg.norm(prev_W_1 - curent_W_1) <= eps_1:
        break

    #print(f"Итерация: {i+1}")
    # print(f"Текущая точка {cur_W}| Следующая точка {next_W}")
    y_proba = sigmoid(X @ curent_W_1)
    y_class = np.where(y_proba >= 0.5, 1, 0)
    accuracy = (y_class == y).sum() / len(y)
    #accuracy = accuracy_score(y, y_class)
    #print(f"Logloss {logloss(y, y_proba)}")
    #print(f"Accuracy {accuracy}")
    #print("--------------------------------------------------------")
    results.loc[len(results.index)] = ['Adam', accuracy, i+1]
```

# **1. Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы**

```python
import pandas as pd
```

```python
df_ratings = pd.read_csv("ratings.csv", sep = '\t')
df_ratings.head()
```

```python
df_movies = pd.read_csv("movies.csv", sep = '|', encoding = 'ISO-8859-1')
df_movies.head()
```

```python
df_movies = df_movies.drop(['Unnamed: 3'], axis=1)
```

```python
df_movies.head()
```

# **2. Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок**

```python
df_ratings[['User_id', 'Rating']].groupby(['User_id']).count().sort_values(by='Rating', ascending=False).head(1)
```

# **3. Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь**

```python
df = df_ratings.loc[df_ratings['User_id'] == 405]
df.head()
```

# **4. Добавьте к датафрейму из задания 3 столбцы:**
**- По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет**

**- Cтолбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей**

```python
df_movies.info()
```

```python
df1 = df.merge(df_movies, how='left', on='Movie_id')
```

```python
df1 = df1.drop(['Movie_title_release_date', 'Video_release_date', 'IMDb_URL'], axis=1)
df1.head()
```

```python
df_ratings_1 = df_ratings[['Movie_id', 'Rating']].groupby(['Movie_id']).count()
df_ratings_1.head(5)
```

```python
df_ratings_1.rename(columns = {'Rating' : 'Rating_count'}, inplace = True)
df_ratings_1.head(5)
```

```python
df_ratings_2 = df_ratings[['Movie_id', 'Rating']].groupby(['Movie_id']).mean()
df_ratings_2.rename(columns = {'Rating' : 'Total_rating'}, inplace = True)
df_ratings_2.head(5)
```

```python
df2 = df1.merge(df_ratings_1, how='left', on='Movie_id').merge(df_ratings_2, how='left', on='Movie_id')
```

```python
df2.head()
```

# **5. Сформируйте X_train, X_test, y_train, y_test**

```python
from sklearn.model_selection import train_test_split
```

```python
df2.info()
```

```python
X, y = df2[['Unknown',
            'Action',
            'Adventure',
            'Animation',
            "Children's",
            'Comedy',
            'Crime',
            'Documentary',
            'Drama',
            'Fantasy',
            'Film_Noir',
            'Horror',
            'Musical',
            'Mystery',
            'Romance',
            'Sci_Fi',
            'Thriller',
            'War',
            'Western',
            'Rating_count',
            'Total_rating']], df2['Rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# **6. Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

# **7. Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии**

```python
from sklearn.metrics import accuracy_score
```

```python
# Оцениваем качество на выборке для обучения
accuracy_score(y_train, model.predict(X_train))
```

```python
# Оцениваем качество на выборке для тестирования
accuracy_score(y_test, model.predict(X_test))
```

# **10. Посчитайте средствами спарка среднюю оценку для каждого жанра**

```python
df_spark_joined = df_ratings_spark.join(df_movies_spark, 'Movie_id', how = 'left')
df_spark_joined.show()
```

```python
df_spark_joined.dtypes
```

```python
import pyspark.sql.functions as F
```

```python
df_spark_Unknown = df_spark_joined.groupBy('Unknown').avg('Rating').where(df_spark_joined['Unknown'] == 1).withColumn('Unknown', F.regexp_replace('Unknown', '1', 'Unknown')).withColumnRenamed('Unknown', 'Genre')
df_spark_Action = df_spark_joined.groupBy('Action').avg('Rating').where(df_spark_joined['Action'] == 1).withColumn('Action', F.regexp_replace('Action', '1', 'Action')).withColumnRenamed('Action', 'Genre')
df_spark_Adventure = df_spark_joined.groupBy('Adventure').avg('Rating').where(df_spark_joined['Adventure'] == 1).withColumn('Adventure', F.regexp_replace('Adventure', '1', 'Adventure')).withColumnRenamed('Adventure', 'Genre')
df_spark_Animation = df_spark_joined.groupBy('Animation').avg('Rating').where(df_spark_joined['Animation'] == 1).withColumn('Animation', F.regexp_replace('Animation', '1', 'Animation')).withColumnRenamed('Animation', 'Genre')
df_spark_Children_s = df_spark_joined.groupBy("Children's").avg('Rating').where(df_spark_joined["Children's"] == 1).withColumn("Children's", F.regexp_replace("Children's", '1', 'Children_s')).withColumnRenamed("Children's", 'Genre')
df_spark_Comedy = df_spark_joined.groupBy('Comedy').avg('Rating').where(df_spark_joined['Comedy'] == 1).withColumn('Comedy', F.regexp_replace('Comedy', '1', 'Comedy')).withColumnRenamed('Comedy', 'Genre')
df_spark_Crime = df_spark_joined.groupBy('Crime').avg('Rating').where(df_spark_joined['Crime'] == 1).withColumn('Crime', F.regexp_replace('Crime', '1', 'Crime')).withColumnRenamed('Crime', 'Genre')
df_spark_Documentary = df_spark_joined.groupBy('Documentary').avg('Rating').where(df_spark_joined['Documentary'] == 1).withColumn('Documentary', F.regexp_replace('Documentary', '1', 'Documentary')).withColumnRenamed('Documentary', 'Genre')
df_spark_Drama = df_spark_joined.groupBy('Drama').avg('Rating').where(df_spark_joined['Drama'] == 1).withColumn('Drama', F.regexp_replace('Drama', '1', 'Drama')).withColumnRenamed('Drama', 'Genre')
df_spark_Fantasy = df_spark_joined.groupBy('Fantasy').avg('Rating').where(df_spark_joined['Fantasy'] == 1).withColumn('Fantasy', F.regexp_replace('Fantasy', '1', 'Fantasy')).withColumnRenamed('Fantasy', 'Genre')
df_spark_Film_Noir = df_spark_joined.groupBy('Film_Noir').avg('Rating').where(df_spark_joined['Film_Noir'] == 1).withColumn('Film_Noir', F.regexp_replace('Film_Noir', '1', 'Film_Noir')).withColumnRenamed('Film_Noir', 'Genre')
df_spark_Horror = df_spark_joined.groupBy('Horror').avg('Rating').where(df_spark_joined['Horror'] == 1).withColumn('Horror', F.regexp_replace('Horror', '1', 'Horror')).withColumnRenamed('Horror', 'Genre')
df_spark_Musical = df_spark_joined.groupBy('Musical').avg('Rating').where(df_spark_joined['Musical'] == 1).withColumn('Musical', F.regexp_replace('Musical', '1', 'Musical')).withColumnRenamed('Musical', 'Genre')
df_spark_Mystery = df_spark_joined.groupBy('Mystery').avg('Rating').where(df_spark_joined['Mystery'] == 1).withColumn('Mystery', F.regexp_replace('Mystery', '1', 'Mystery')).withColumnRenamed('Mystery', 'Genre')
df_spark_Romance = df_spark_joined.groupBy('Romance').avg('Rating').where(df_spark_joined['Romance'] == 1).withColumn('Romance', F.regexp_replace('Romance', '1', 'Romance')).withColumnRenamed('Romance', 'Genre')
df_spark_Sci_Fi = df_spark_joined.groupBy('Sci_Fi').avg('Rating').where(df_spark_joined['Sci_Fi'] == 1).withColumn('Sci_Fi', F.regexp_replace('Sci_Fi', '1', 'Sci_Fi')).withColumnRenamed('Sci_Fi', 'Genre')
df_spark_Thriller = df_spark_joined.groupBy('Thriller').avg('Rating').where(df_spark_joined['Thriller'] == 1).withColumn('Thriller', F.regexp_replace('Thriller', '1', 'Thriller')).withColumnRenamed('Thriller', 'Genre')
df_spark_War = df_spark_joined.groupBy('War').avg('Rating').where(df_spark_joined['War'] == 1).withColumn('War', F.regexp_replace('War', '1', 'War')).withColumnRenamed('War', 'Genre')
df_spark_Western = df_spark_joined.groupBy('Western').avg('Rating').where(df_spark_joined['Western'] == 1).withColumn('Western', F.regexp_replace('Western', '1', 'Western')).withColumnRenamed('Western', 'Genre')
```

```python
from functools import reduce
from pyspark.sql import DataFrame

df_spark_1 = [df_spark_Unknown, df_spark_Action, df_spark_Adventure,
              df_spark_Animation, df_spark_Children_s, df_spark_Comedy,
              df_spark_Crime, df_spark_Documentary, df_spark_Drama,
              df_spark_Fantasy, df_spark_Film_Noir, df_spark_Horror,
              df_spark_Musical, df_spark_Mystery, df_spark_Romance,
              df_spark_Sci_Fi, df_spark_Thriller, df_spark_War,
              df_spark_Western]
df_spark_genres = reduce(DataFrame.unionAll, df_spark_1)
```
